Task 1 - Give 2 examples of all string functions which should be meaningful

```python
#capitalize()	Converts the first character to upper case
user_name = "manoj kumar"
print(user_name.capitalize())

car_name = "suzuki swift"
print(car_name.capitalize())
```

```python
# casefold()	Converts string into lower case

city_name = "NEW DELHI"
print(city_name.casefold())

town_name = "SALEM"
print(town_name.casefold())
```

```python
# center()	Returns a centered string
word = "course"
print(word.center(20,'-'))

word_1 = "automatic"
print(word_1.center(10,"_"))
```

```python
# count()	Returns the number of times a specified value occurs in a string
country_name = 'Australia'
print(country_name.count('a'))

city_name = "Canberra"
print(city_name.count('r'))
```

```python
# encode()	Returns an encoded version of the string
given_name = "My name is Månoj Kumar"
print(given_name.encode(encoding='ascii',errors='replace'))

firs_name = "My name is Råm Kumar"
print(given_name.encode(encoding='ascii',errors='ignore'))
```

```python
# endswith()	Returns true if the string ends with the specified value
place_name = "Brisbane"
print(place_name.endswith('a'))

village = "Korattur"
print(village.endswith("r"))
```

```python
# expandtabs()	Sets the tab size of the string
car_model = "H\to\tn\td\ta"
print(car_model.expandtabs())

fuel = "P\te\tt\tr\to\tl\t"
print(fuel.expandtabs())
```

```python
# find()	Searches the string for a specified value and returns the position of where it was found
sentence = "The mountain is huge and looks massive"
print(sentence.find("huge"))

hospital = "Velammal"
print(hospital.find("V"))
```

```python
# format()	Formats specified values in a string
print("My name is {}, and I am from {}".format("Manoj","Chennai"))
print("I find this {}".format("Interesting"))
```

```python
# format_map()	Formats specified values from a dictionary in a string
my_detail = {"name" : "MK", "age" : 26}
sample_sentence = "Happy birthday {name} you are now on level {age}!"
print(sample_sentence.format_map(my_detail))

student_detail = {"name" : "MK", "course" : "DS-ML-AI"}
sample_sentence_1 = "{name} is enrolled on {course} course"
print(sample_sentence_1.format_map(student_detail))
```

```python
# index()	Searches the string for a specified value and returns the position of where it was found
student_subject_review = "Maths is difficult"
print(student_subject_review.index("u"))

mobile = "samsung"
print(mobile.index("g"))
```

```python
# isalnum()	Returns True if all characters in the string are alphanumeric
string = "Company"
print(string.isalnum())

string_1 = "456 "
print(string_1.isalnum())
```
