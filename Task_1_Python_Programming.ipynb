Task 1 - Give 2 examples of all string functions which should be meaningful

```python
#capitalize()	Converts the first character to upper case
user_name = "manoj kumar"
print(user_name.capitalize())

car_name = "suzuki swift"
print(car_name.capitalize())
```

```python
# casefold()	Converts string into lower case

city_name = "NEW DELHI"
print(city_name.casefold())

town_name = "SALEM"
print(town_name.casefold())
```

```python
# center()	Returns a centered string
word = "course"
print(word.center(20,'-'))

word_1 = "automatic"
print(word_1.center(10,"_"))
```

```python
# count()	Returns the number of times a specified value occurs in a string
country_name = 'Australia'
print(country_name.count('a'))

city_name = "Canberra"
print(city_name.count('r'))
```

```python
# encode()	Returns an encoded version of the string
given_name = "My name is Månoj Kumar"
print(given_name.encode(encoding='ascii',errors='replace'))

firs_name = "My name is Råm Kumar"
print(given_name.encode(encoding='ascii',errors='ignore'))
```

```python
# endswith()	Returns true if the string ends with the specified value
place_name = "Brisbane"
print(place_name.endswith('a'))

village = "Korattur"
print(village.endswith("r"))
```

```python
# expandtabs()	Sets the tab size of the string
car_model = "H\to\tn\td\ta"
print(car_model.expandtabs())

fuel = "P\te\tt\tr\to\tl\t"
print(fuel.expandtabs())
```

```python
# find()	Searches the string for a specified value and returns the position of where it was found
sentence = "The mountain is huge and looks massive"
print(sentence.find("huge"))

hospital = "Velammal"
print(hospital.find("V"))
```

```python
# format()	Formats specified values in a string
print("My name is {}, and I am from {}".format("Manoj","Chennai"))
print("I find this {}".format("Interesting"))
```

```python
# format_map()	Formats specified values from a dictionary in a string
my_detail = {"name" : "MK", "age" : 26}
sample_sentence = "Happy birthday {name} you are now on level {age}!"
print(sample_sentence.format_map(my_detail))

student_detail = {"name" : "MK", "course" : "DS-ML-AI"}
sample_sentence_1 = "{name} is enrolled on {course} course"
print(sample_sentence_1.format_map(student_detail))
```

```python
# index()	Searches the string for a specified value and returns the position of where it was found
student_subject_review = "Maths is difficult"
print(student_subject_review.index("u"))

mobile = "samsung"
print(mobile.index("g"))
```

```python
# isalnum()	Returns True if all characters in the string are alphanumeric
string = "Company"
print(string.isalnum())

string_1 = "456 "
print(string_1.isalnum())
```

```python
# isalpha()	Returns True if all characters in the string are in the alphabet
city = "Area 52"
print(city.isalpha())

country = "America"
print(country.isalpha())
```

```python
# isascii()	Returns True if all characters in the string are ascii characters
name = "Manoj"
print(name.isascii())

score = "321"
print(score.isascii())
```

```python
# isdecimal()	Returns True if all characters in the string are decimals
score = "123"
print(score.isdecimal())

point = "5.6"
print(point.isdecimal())
```

```python
# isdigit()	Returns True if all characters in the string are digits
points = "34678"
print(points.isdigit())

player = "Dhoni"
print(player.isdigit())
```

```python
# isidentifier()	Returns True if the string is an identifier
security_key = "Password"
print(security_key.isidentifier())

password = "1234password"
print(password.isidentifier())
```

```python
# islower()	Returns True if all characters in the string are lower case
state_name = "TAMILNADU"
print(state_name.islower())

country = "argentina"
print(country.islower())
```

```python
# isnumeric()	Returns True if all characters in the string are numeric
bus_no = "25"
print(bus_no.isnumeric())

train_no = "SL4567"
print(train_no.isnumeric())
```

```python
# isprintable()	Returns True if all characters in the string are printable
string = "Roshan Arts"
print(string.isprintable())

string_1 = "Roshan #B\n"
print(string_1.isprintable())
```

```python
# isspace()	Returns True if all characters in the string are whitespaces
shopping_mall = 'Spencer   Plaza   '
print(shopping_mall.isspace())

book = "         "
print(book.isspace())
```

```python
# istitle()	Returns True if the string follows the rules of a title
statement = "I Am Interested On Cars"
print(statement.istitle())

statement_1 = "are you feeling ok?"
print(statement_1.istitle())
```

```python
# isupper()	Returns True if all characters in the string are upper case
state_name = "WESTERN GHATS"
print(state_name.isupper())

place_name = "madurai"
print(place_name.isupper())
```

```python
# join()	Converts the elements of an iterable into a string
tuple = ("M","A","N","O","J")
print("".join(tuple))

tuple1 = ("K","U","M","A","R")
print("".join(tuple1))
```

```python
# ljust()	Returns a left justified version of the string
fruit = "banana"
print(fruit.ljust(10, "O"))

name = "igl"
print(name.ljust(5,"o"))
```

```python
# lower()	Converts a string into lower case
state_name = "DHARMAPURI"
print(state_name.lower())

country = "JAPAN"
print(country.lower())
```

```python
# lstrip()	Returns a left trim version of the string
name = "     madurai"
print(name.lstrip())

word = "   important"
print(word.lstrip())
```

```python
# maketrans()	Returns a translation table to be used in translations
statement = "Hello All!"
modified_statement = str.maketrans("H", "C")
print(statement.translate(modified_statement))

statement_1 = "Congratulations"
modified_statement_1 = str.maketrans("s", " ")
print(statement_1.translate(modified_statement_1))
```

```python
# partition()	Returns a tuple where the string is parted into three parts
sentence = "I practice coding 1 hour per day"
print(sentence.partition("coding"))

highways = "NH India has longest roadways"
print(highways.partition("has"))
```

```python
# replace()	Returns a string where a specified value is replaced with a specified value
destination = "Ooty has 14 hairpin bends"
print(destination.replace("14","26"))

destination_1 = "Coimbatore is usually hot"
print(destination_1.replace("Coimbatore","Chennai"))
```

```python
# rfind()	Searches the string for a specified value and returns the last position of where it was found
manufacturer = "Honeywell"
print(manufacturer.rfind("l"))

manufacturer_1 = "Ramraj"
print(manufacturer_1.rfind("R"))
```
