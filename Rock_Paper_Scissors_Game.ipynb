Objective:
Build a simple Rock-Paper-Scissors game where the player competes with the computer.
 This task will test your understanding of:
●	Input/output
●	Conditional logic
●	Random module
●	Loops
●	Exception handling

Instructions:
######1.	The user should enter their choice: rock, paper, or scissors.
######2.	The computer should randomly select one of these.
######3.	Compare both and print who wins.
######4.	Ask if the player wants to play again- if yes, repeat; if no, exit.
######5.	Handle invalid inputs gracefully using try-except or condition checks.

```python
import random

def rock_paper_scissors():

  options = ["Rock","Paper","Scissors"]

  while True:
    try:
      sys_choice = random.choice(options)
      user_input = input("Enter your input:")
      if user_input not in options:
          print("Invalid input. Enter Rock,Paper,Scissors")
      elif user_input == "Rock" and sys_choice == "Rock":
          print("It's a tie")
      elif user_input == "Scissors" and sys_choice == "Scissors":
          print("It's a tie")
      elif user_input == "Paper" and sys_choice == "Paper":
          print("It's a tie")
      elif user_input == "Rock" and sys_choice == "Paper":
          print("System won")
      elif user_input == "Paper" and sys_choice == "Scissors":
          print("System won")
      elif user_input == "Scissors" and sys_choice == "Rock":
          print("System won")
      elif user_input == "Rock" and sys_choice == "Scissors":
          print("User won")
      elif user_input == "Scissors" and sys_choice == "Paper":
          print("User won")
      elif user_input == "Paper" and sys_choice == "Rock":
          print("User won")

      play_again = input("Do you want to play again (yes/no):")

      if play_again == "no":
        print("Thanks for playing! See you next time!")
        break
      elif play_again == "yes":
        rock_paper_scissors()
        #print("Invalid response. Exiting game.")
        #break

    except KeyboardInterrupt:
       print("Game interrupted. Goodbye!")
       break

rock_paper_scissors()
```
