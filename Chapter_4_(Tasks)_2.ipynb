```python
import pandas as pd
```

```python
# Step 1: Load and combine the data to one - sales_all
import pandas as pd

sales_2017 = pd.read_csv(filepath_or_buffer="/content/drive/MyDrive/Python - Data Analysis/Sales Transactions-2017.csv")
sales_2018 = pd.read_csv(filepath_or_buffer="/content/drive/MyDrive/Python - Data Analysis/Sales Transactions-2018.csv")
sales_2019 = pd.read_csv(filepath_or_buffer="/content/drive/MyDrive/Python - Data Analysis/Sales Transactions-2019.csv")

sales_all = pd.concat(objs=[sales_2017,sales_2018,sales_2019])
sales_all
```

```python
# Step 2: Shape of uncleaned data
sales_all.shape
```

```python
# Step 3: Remove blank rows

sales_cleaned_df = sales_all.dropna(subset=['Date', 'Voucher', 'Party', 'Product'], how='all')

sales_cleaned_df.shape
```

```python
# Step 4: Remove total rows
sales_cleaned_df = sales_cleaned_df[~sales_cleaned_df['Voucher'].astype(str).str.contains('Total', na=False)]
sales_cleaned_df = sales_cleaned_df[~sales_cleaned_df['Party'].astype(str).str.contains('Total', na=False)]

sales_cleaned_df.shape
```

```python
# Step 5: Clean number columns

numeric_column = ['Qty', 'Rate', 'Gross', 'Disc', 'Voucher Amount']

sales_cleaned_df[numeric_column] = sales_cleaned_df[numeric_column].apply(lambda x: pd.to_numeric(x.astype(str).str.replace(',',''),errors='coerce'))
```

```python
# Step 6: Convert date to date type

sales_cleaned_df['Date'] = pd.to_datetime(sales_cleaned_df['Date'],dayfirst=True,errors='coerce')

sales_cleaned_df = sales_cleaned_df.dropna(subset=['Date'])

#sales_cleaned_df.dtypes

sales_cleaned_df #After dropping the dates, we get the required number of rows as shown in cleaned dataset
```

```python
# Step 7: Drop the unwanted columns

sales_cleaned_df = sales_cleaned_df.drop(columns=['Gross','Disc','Voucher Amount'])

sales_cleaned_df
```

```python
# Step 8: Drop the index column

sales_cleaned_df.drop(sales_cleaned_df.columns[0],axis=1)

sales_cleaned_df
```

#### Task 4: Lambda function 2 example with Map, Filter, Reduce

###### Map

```python
numbers = [1, 2, 3, 4, 5]

result = list(map(lambda x: x * 2, numbers))

print(result)
```

```python
names = ['Arun', 'Babu', 'Chandru']

result = list(map(lambda x: x.upper(), names))

print(result)
```

###### Filter

```python
numbers = [1, 2, 3, 4, 5, 6]

result = list(filter(lambda x: x % 2 == 0, numbers))

print(result)
```

```python
names = ['Arun', 'Babu', 'Chandru']

result = list(filter(lambda x: len(x) > 4, names))

print(result)
```

###### Reduce


```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

result = reduce(lambda x, y: x + y, numbers)

print(result)
```

```python
from functools import reduce

numbers = [3, 7, 1, 9, 4]

result = reduce(lambda x, y: x if x > y else y, numbers)

print(result)
```
