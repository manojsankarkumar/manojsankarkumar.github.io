## Task 2: Create 3 DataFrames using Dictionaries

###### Dataframe using List

```python
import pandas as pd

employee_info = {"Employee Name": ["Jack","John","Abdul","Mohan","Arvind","Ravi","Prabhu","Praveen"],
                 "ID":[1000,1001,1002,1003,1004,1005,1006,1007],
                 "Department":["Dev","Dev","Dev","Testing","Testing","HR","HR","HR"],
                 "Experience (Yrs)": [10,8,3,14,5,9,2,4],
                 "Location":["Chennai","Pune",'Delhi','Mumbai',"Gujarat",'Bangalore',"Jaipur","Assam"],
                 "Shift Period":["Day","Night","Day","Night","Day","Night","Day","Night",]}

pd.DataFrame(employee_info)
```

######Dataframe using Tuple

```python
import pandas as pd

# 1. List of tuples (most common)
data = [("Aakash", 25, 88.5),
        ("Babu",   30, 92.0),
        ("Chandru", 35, 79.3)]

df_all = pd.DataFrame(data, columns=["Name", "Age", "Score"])

print(df_all)

# 2. Single tuple of tuples
data = (("Aakash", 25), ("Babu", 30), ("Chandru", 35))
df_single = pd.DataFrame(data, columns=["Name", "Age"])

print(df_single)

# 3. Named tuples → column names are inferred automatically!
from collections import namedtuple

Person = namedtuple("Person", ["Name", "Age", "Score"])
data = [Person("Aakash", 25, 88.5),
        Person("Babu",   30, 92.0)]

df_double = pd.DataFrame(data)   # no need to pass columns!
print(df_double)

# 4. Tuple as a single row

print("Tuple as a single row")
row = ("Aakash", 25, 88.5)
df_single_row = pd.DataFrame([row], columns=["Name", "Age", "Score"])

print(df_single_row)
```

###### Dataframe using Set

```python
# ⚠️ Sets are UNORDERED — column/row order is not guaranteed.

# 1. List of sets (each set = one row)
data = [{"Arvind", 25, 88.5},
        {"Bani",   30, 92.0}]

df = pd.DataFrame(data)   # column names are the set values (unpredictable!)
# Not recommended — sets have no keys, so columns are arbitrary
print(df)


# 2. Set as index values
my_set = {"Arvind", "Bani", "Chandru"}
df = pd.DataFrame({"score": [88, 92, 79]}, index=list(my_set))
# ⚠️ Order not guaranteed since sets are unordered
print(df)

# 3. Set of tuples → convert to list first
data = {("Arvind", 25), ("Bani", 30), ("Chandru", 35)}
df = pd.DataFrame(list(data), columns=["name", "age"])
# ⚠️ Row order not guaranteed
print(df)

# ── SUMMARY TABLE ────────────────────────────────────────────────────────────
#summary = pd.DataFrame([
#    ("List of tuples",      "✅ Recommended",   "Preserves order"),
#    ("Tuple of tuples",     "✅ Works fine",    "Preserves order"),
#    ("Named tuples",        "✅ Best practice", "Auto-infers column names"),
#    ("List of sets",        "⚠️ Avoid",         "No keys, unpredictable columns"),
#    ("Set of tuples",       "⚠️ Careful",       "Row order not guaranteed"),
#    ("Set as index",        "⚠️ Careful",       "Index order not guaranteed"),
#], columns=["Method", "Recommendation", "Reason"])"""
```
