#### 1.3 Pull up the average insurance charges based on region and sex

### Discrete and Continuous data

* Discrete - Just countable / can't be Segmented
* Continuous - Measurable/Have units/Can also be segmented

```python
insurance_data.groupby(by=["region","sex"])["charges"].mean().round(2).sort_values(ascending=False) #Chain method
```

## 2. Pivot table

```python
import pandas as pd

# Load the data first
insurance_df = pd.read_csv("/content/drive/MyDrive/Python - Data Analysis/Insurance.csv")

pd.pivot_table(data=insurance_df,values=["charges"],index="region",columns=["sex","children"]).round(2) # presentable comparing 2 categorical features vs 1 continuous data
```

## 3. Aggregate

```python
import pandas as pd

# Load the data first
insurance_df = pd.read_csv("/content/drive/MyDrive/Python - Data Analysis/Insurance.csv")
insurance_df.head()
```

##### 3.1 Find out the average insurance charges based on Region

```python
insurance_df.groupby(by=["region","sex"])["charges"].mean().round(2).sort_values(ascending=False) #Chain method
```

```python
insurance_df.groupby(by=["region","sex"])["charges"].agg(func=["mean","min","max","sum"]).round()
```

## 5. Concat

## 6. Apply

```python
import pandas as pd

# Load the data first
insurance_df = pd.read_csv("/content/drive/MyDrive/Python - Data Analysis/Insurance.csv")
insurance_df.head()
```

```python
insurance_df.describe()
```

#### 6.1 Create a new column in the same dataframe with the name 'Eligibility Criteria' where if age is > 20, say Eligible else not

```python
def get_age(x):
  if x > 20:
    return "Eligible"
  else:
    return "Not eligible"
```

```python
insurance_df["Eligibility Criteria"] = insurance_df['age'].apply(get_age)
insurance_df
```

## 7. Lambda

* Anonymous function italicized text/ nameless function

```python
insurance_df
```

```python
insurance_df["Eligibility Criteria_1"] = insurance_df['age'].apply(lambda x:1 if x > 20 else 0)
insurance_df
```

## 8. Del

```python
del insurance_df["Eligibility Criteria_1"]
insurance_df
```
