Objective:
Create a word-guessing game where the player tries to guess the letters of a hidden word.

This will test your:
●	Loops
●	Conditional logic
●	String manipulation
●	Exception handling
●	Logical thinking

Instructions:
######1.	Create a list of random words (like “python”, “machine”, “learning”, “bridge”).
######2.	Randomly select one word for the game.
######3.	Show the word as blanks (_ _ _ _) and let the player guess one letter at a time.
######4.	Keep track of wrong guesses (max 6 attempts).
######5.	End the game when the player either guesses the full word or uses all chances.



```python
import random

def mini_hangman():

  words = ["python","machine","learning","bridge"]
  random_word = random.choice(words)

  blanks = len(random_word) * ["_"]

  guess_left = []
  max_attempts = 6

  while True:
      print("\nWord:"," ".join(blanks))
      print("Wrong guesses:", guess_left)
      print("Attempts_remaining", max_attempts - len(guess_left))

      guess = input("Enter a guess letter:")

      if len(guess) != 1 or not guess.isalpha():
        print("Please enter single letter only")
        continue

      if guess in guess_left or guess in blanks:
        print("You guessed that letter! Enter different one")
        continue

      if guess in random_word:
        print("Correct")
        for i in range(len(random_word)):
          if random_word[i] == guess:
            blanks[i] = guess

      else:
        print("Wrong")
        guess_left.append(guess)

      if "_" not in blanks:
        print("You win! The word is",random_word)
        break

      if len(guess_left) >= max_attempts:
        print("Game over! The word was:", random_word)
        break

mini_hangman()
```
