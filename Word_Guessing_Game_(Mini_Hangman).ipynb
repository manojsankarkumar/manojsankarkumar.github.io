```python
import random

def word_guessing_game():

    words = ["python", "machine", "learning", "bridge"]
    secret_word = random.choice(words)

    blanks = ["_"] * len(secret_word)

    guesses_left = []
    max_attempts = 6

    while True:

        print("\nWord:", " ".join(blanks))
        print("Wrong guesses:", guesses_left)
        print("Attempts remaining:", max_attempts - len(guesses_left))

        guess = input("Guess a letter: ")

        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a single letter only.")
            continue

        if guess in guesses_left or guess in blanks:
            print("You already guessed that letter! Try a different one.")
            continue

        if guess in secret_word:
            print("Correct!")
            for i in range(len(secret_word)):
                if secret_word[i] == guess:
                    blanks[i] = guess
        else:
            print("Wrong!")
            guesses_left.append(guess)

        if "_" not in blanks:
            print("\nYou won! The word was:", secret_word)
            break

        if len(guesses_left) >= max_attempts:
            print("\nGame over! The word was:", secret_word)
            break

word_guessing_game()
```
