```python
#Positional Arguments
def get_details(name,place,expected_price):
  print("Client name:",name)
  print("Place:", place)
  print("Expected price:",expected_price)

get_details(place="Chennai",expected_price="97.5",name='Manoj')
```

```python
#Default arguments
def get_details_chennai(name,expected_price,place="Chennai"):
  print("Client name:",name)
  print("Place:", place)
  print("Expected price:",expected_price)

get_details_chennai(expected_price=95.5,name="Manoj Kumar")
```

```python
#Variable-length argument
def get_details_chennai(name,place,*expected_price):
  print("Client name:",name)
  print("Place:", place)
  print("Expected price:",expected_price)

get_details_chennai("Manoj Kumar","Chennai",95.5,90,67)
```

```python
#Keyword argument
def get_details_chennai(name,place,**expected_price):
  print("Client name:",name)
  print("Place:", place)
  print("Expected price:",expected_price)

get_details_chennai("Manoj Kumar","Chennai",ask_1=95.5,ask_2=90,ask_3=67)
```

```python
#5.4 Challenge 2 - Find the factors of a number
user_input = int(input("Enter your number"))

for i in range(1,user_input+1):
  if user_input % i == 0:
    print(i)
```

```python
user_input = int(input("Enter your number"))

result = []
for i in range(1,user_input+1):
  if user_input % i == 0:
    result.append(i)
print("The factors of {} is {}".format(user_input,result))
```
