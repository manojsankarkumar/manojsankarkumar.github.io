```python
#2.3 Manipulate strings with string methods
learner_1 = 'Manoj'
learner_2 = '   Kumar'
learner_3 = 'Toyota   '
learner_4 = '   Honda   '

print(learner_1,learner_2,learner_3,learner_4)

print(learner_1)
print(learner_2)
print(learner_3)
print(learner_4)
```

```python
learner_1 = learner_1.upper()
learner_2 = learner_2.lstrip()
learner_3 = learner_3.rstrip()
learner_4 = learner_4.strip()

print(learner_1)
print(learner_2)
print(learner_3)
print(learner_4)
```
