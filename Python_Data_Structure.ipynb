1. List

* Accepts duplicate values
* One dimensional
* Heterogeneous structure - can hold more than one data type
* Indexing and slicing is possible
* Mutable - values once assigned can be changed in the future

```python
empty_cart = []
```

```python
empty_cart
```

```python
type(empty_cart)
```

```python
item_to_buy = ["Apple", "Orange", "Grapes", "Pomogrenate","Apple",6.76,8,9]
item_to_buy
```

```python
type(item_to_buy)
```

```python
len(item_to_buy)
```

```python
item_to_buy[3] #indexing
```

```python
item_to_buy[-5] #negative indexing
```

```python
item_to_buy[2:5] #positive indexing
```

```python
item_to_buy[-6:-3] #negative indexing
```

```python
item_to_buy
```

```python
item_to_buy[1] = "watermelon"
```

```python
item_to_buy
```

2. Set

* Accepts only unique values, no duplicates
* One dimensional
* Heterogeneous structure - can hold more than one data type
* Indexing and slicing not possible
* Mutable

```python
empty_cart_2 = set()
```

```python
type(empty_cart_2)
```

```python
item_to_buy_2 = {"Apple", "Orange", "Grapes", "Pomogrenate","Apple",6.76,8,9}
item_to_buy_2
```

```python
type(item_to_buy_2)
```

```python
item_to_buy_2.add("banana")
```

```python
item_to_buy_2
```

```python
item_to_buy_2.discard("Apple")
```

```python
item_to_buy_2
```

3. Tuple

* Accepts only unique values, no duplicates
* One dimensional
* Heterogeneous structure - can hold more than one data type
* Indexing and slicing possible
* Immutable

```python
empty_cart_3 = ()
```

```python
type(empty_cart_3)
```

```python
item_to_buy_3 = ("Apple", "Orange", "Grapes", "Pomogrenate","Apple",6.76,8,9)
item_to_buy_3
```

```python
item_to_buy_3[0:3]
```

```python
item_to_buy_3[0] = "Sugarcane"
```

```python
item_to_buy_3.count('Apple')
```

```python
item_to_buy_3.index("Grapes")
```

4. Dictionary

* It is a Key:Value pair
* Two dimensional
* Heterogeneous
* Mutable
* Indexing and slicing is not possible
* Dont create duplicate key, but can create duplicate values

```python
empty_cart_4 = {}
```

```python
type(empty_cart_4)
```

```python
item_to_buy_4 = {"Potatoe":5,"Tomato":5,"Apple":5}
item_to_buy_4
```

```python
type(item_to_buy_4)
```

```python
item_to_buy_4.clear()
```

```python
item_to_buy_4.keys()
```

```python
item_to_buy_4.values()
```
