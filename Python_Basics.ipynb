```python
## Level 3. Give multiple conditions
#1. User should enter the number between 1-100. If it's entered wrong, the program should say, "Invalid Number"
#2. If the user guessed number > secret number, then program should say, "Your number is greater than secret number"
#3. If the user guessed number < secret number, then program should say, "Your number is lesser than secret number"


print("Welcome to Number Guessing Game!!!")
print('Guess the number between 1 and 100')

secret_number = 98

for i in range(0,7):
  user_input = int(input("Enter the number:"))
  if user_input < 1 or user_input > 100:
    print("Invalid number! Enter the number between 1 and 100")
  elif user_input > secret_number:
    print("Your number is higher")
  elif user_input < secret_number:
    print("Your number is lesser")
  else:
    print("Your guessed number is correct!")
    print(f"Congratulations!, you have guessed the number {secret_number} in {i} attempts")
    break
```

```python
## 6. How to handle invalid characters or values
import random

print("Welcome to Number Guessing Game!!!")
print('Guess the number between 1 and 100')

secret_number = random.randint(1,100)

for i in range(0,7):
    try:
      user_input = int(input("Enter the number:"))
      if user_input < 1 or user_input > 100:
        print("Invalid number! Enter the number between 1 and 100")
      elif user_input > secret_number:
        print("Your number is higher")
      elif user_input < secret_number:
        print("Your number is lesser")
      else:
        print("Your guessed number is correct!")
        break
    except ValueError:
      print("Invalid characters. Please enter a number")


if secret_number == user_input:
   print(f"Congratulations!, you have guessed the number {secret_number} in {i} attempts")
else:
  print("All your trials is exceeded. Try Again!!")
```

```python
## 7. Take this program to run as a function when called
def number_guessing_game():
  import random

  print("Welcome to Number Guessing Game!!!")
  print('Guess the number between 1 and 100')

  secret_number = random.randint(1,100)

  for i in range(0,7):
      try:
        user_input = int(input("Enter the number:"))
        if user_input < 1 or user_input > 100:
          print("Invalid number! Enter the number between 1 and 100")
        elif user_input > secret_number:
          print("Your number is higher")
        elif user_input < secret_number:
          print("Your number is lesser")
        else:
          print("Your guessed number is correct!")
          break
      except ValueError:
        print("Invalid characters. Please enter a number")


  if secret_number == user_input:
    print(f"Congratulations!, you have guessed the number {secret_number} in {i} attempts")
  else:
    print("All your trials is exceeded. Try Again!!")
```
