```python
#1.	while loop
print("Printing numbers from 1 to 10")
num = 1
while num < 11:
  print(num)
  num = num + 1

#To print multiples of 2
print("Multiples of 2")
number = 0
while number < 20:
  number = number + 2
  print(number)
```

```python
#2.	for loop
print("Looping over a list")
languages = ['Tamil', 'English', 'Kannada', 'Hindi', 'French', "Sanskrit"]
for i in languages:
    print(i)

print("Looping over a string")
language = 'Python'
for c in language:
    print(c)
```

```python
#3.	if else
input_num = int(input("Enter a number of your choice:"))

if input_num > 0:
  print("The number is positive!")
else:
  print("The number is negative!")
```

```python
#3.	if else
first_num = int(input("Enter your first number:"))
second_num = int(input("Enter your second number:"))

if first_num == second_num:
  print("Equal")
else:
  print("Not equal")
```

```python
#4.	if elif else
total_marks = int(input("Enter your total marks out of 1200: "))

if total_marks > 1200 or total_marks < 0:
    print("Invalid marks! Please enter a value between 0 and 1200.")
elif total_marks >= 1001:
    print("Grade: A")
elif total_marks >= 801:
    print("Grade: B")
elif total_marks >= 601:
    print("Grade: C")
else:
    print("Grade: D")
```

```python
#4.	if elif else
day_of_week = int(input("Enter a weekday number:"))

if day_of_week == 1:
  print("Sunday")
elif day_of_week == 2:
  print("Monday")
elif day_of_week == 3:
  print("Tuesday")
elif day_of_week == 4:
  print("Wednesday")
elif day_of_week == 5:
  print("Thursday")
elif day_of_week == 6:
  print("Friday")
elif day_of_week == 7:
  print("Saturday")
else:
  print("Invalid entry")
```

```python
#5.	user defined functions without for loop, if loop
def day_greeting():
    day_time = int(input("Enter the hour only (0-23): "))

    if day_time > 23 or day_time < 0:
        print("Invalid entry. Please input value between 0 and 23")
    elif day_time >= 6 and day_time <= 12:
        print("Good Morning")
    elif day_time >= 13 and day_time <= 16:
        print("Good Afternoon")
    elif day_time >= 17 and day_time <= 21:
        print("Good Evening")
    else:
        print("Good Night")

day_greeting()
```

```python
#5.	user defined functions without for loop, if loop
def week_output():
    week_day = input("Enter the weekday: ")

    if week_day not in ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]:
        print("Invalid entry. Please enter days of the week")
    elif week_day in ["Monday", "Tuesday"]:
        print("Week beginning")
    elif week_day in ["Wednesday", "Thursday", "Friday"]:
        print("Mid week")
    elif week_day in ["Saturday", "Sunday"]:
        print("End of week")

week_output()
```
